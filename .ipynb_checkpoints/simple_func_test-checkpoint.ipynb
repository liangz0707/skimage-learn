```python
imgL = img_lab[:,:,0]
imgL = skimage.img_as_ubyte(imgL/100)
print imgL.shape,imgL.min(),imgL.max()
#测试一下preserve_range参数
img1 = rescale(imgL,1.0,preserve_range=False)
print img1.shape,img1.min(),img1.max()
imgD = rescale(imgL,0.5,preserve_range=True)
print imgD.shape,imgD.min(),imgD.max()
img2 = rescale(imgL,0.5,preserve_range=False)
print img2.shape,img2.min(),img2.max()
img4 = rescale(imgL,0.25,preserve_range=False)
print img4.shape,img4.min(),img4.max()
img8 = rescale(imgL,0.125,preserve_range=False)
print img8.shape,img8.min(),img8.max()

img1H= rescale(img1,1.0,preserve_range=True)
img2H = rescale(img2,2.0,preserve_range=True)
img4H = rescale(img4,4.0,preserve_range=True)
img8H = rescale(img8,8.0,preserve_range=True)

plt.figure(figsize=(18,16), dpi=80)
plt.subplot(2,2,1)
plt.imshow(img1H,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,2)
plt.imshow(img2H,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,3)
plt.imshow(img4H,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,4)
plt.imshow(img8H,cmap=cm.gray,interpolation='none')
```

```python
def gabor_fn(sigma,theta,Lambda,psi,gamma):
    sigma_x = sigma;
    sigma_y = float(sigma)/gamma;
    
    # Bounding box
    nstds =4;
    xmax = max(abs(nstds*sigma_x*np.cos(theta)),abs(nstds*sigma_y*np.sin(theta)));
    xmax = np.ceil(max(1,xmax));
    ymax = max(abs(nstds*sigma_x*np.sin(theta)),abs(nstds*sigma_y*np.cos(theta)));
    ymax = np.ceil(max(1,ymax));
    xmin = -xmax; ymin = -ymax;
    print ymax,xmax
    (x,y) = np.meshgrid(np.arange(xmin,xmax+1),np.arange(ymin,ymax+1 ));
    (y,x) = np.meshgrid(np.arange(ymin,ymax+1),np.arange(xmin,xmax+1 ));
    print x,y
    # Rotation 
    x_theta=x*np.cos(theta)+y*np.sin(theta);
    y_theta=-x*np.sin(theta)+y*np.cos(theta);
    
    gb= np.exp(-.5*(x_theta**2/sigma_x**2+y_theta**2/sigma_y**2))*np.cos(2*np.pi/Lambda*x_theta+psi);
    return gb
from skimage.filters import gabor_kernel
sigma =4
flt  =  np.real(gabor_kernel(0.1,theta=np.pi/4,sigma_x=sigma, sigma_y=sigma))
# flt = gabor_fn(1,1/4,3,0,1)
print flt.shape

flt= np.round(flt,6)#确定有效的位数。
flt = np.asarray(flt)

plt.figure(figsize=(8,6), dpi=130)
plt.imshow(flt,cmap=cm.gray,interpolation='none')
```

```python
from scipy import ndimage as ndi


f1 = ndi.convolve(img1, flt, mode='wrap')
f2 = ndi.convolve(img2H, flt, mode='wrap')
f4 = ndi.convolve(img4H, flt, mode='wrap')
f8 = ndi.convolve(img8H, flt, mode='wrap')
plt.figure(figsize=(18,16), dpi=80)
plt.subplot(2,2,1)
plt.imshow(f1,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,2)
plt.imshow(f2,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,3)
plt.imshow(f4,cmap=cm.gray,interpolation='none')
plt.subplot(2,2,4)
plt.imshow(f8,cmap=cm.gray,interpolation='none')
```
